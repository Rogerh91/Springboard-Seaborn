## Boxplot

```python
sns.boxplot(titanic.Age.dropna())
```

```python
sns.boxplot(titanic.Age.dropna(), titanic.Sex)
```

```python
sns.boxplot(server.Time)
```

```python
sns.boxplot(mlb.Salary)
```

## Kernel Density Estimate Plot

```python
sns.kdeplot(titanic.Age.dropna(), shade = True)
```

```python
sns.kdeplot(server.Time, shade = True)
```

```python
sns.kdeplot(mlb.Salary, shade = True)
```

```python
sns.distplot(titanic.Age.dropna())
```

```python
sns.distplot(server.Time)
```

```python
sns.distplot(mlb.Salary)
```

## Violin Plot

```python
sns.violinplot(titanic.Age.dropna(), color = "Red", vert = True)
```

```python
sns.violinplot(titanic.Age.dropna(), titanic.Sex)
```

```python
sns.violinplot(mlb.Salary, vert = True)
```

## Cumulative distribution function

```python
sns.kdeplot(titanic.Age.dropna(), cumulative = True)
```

```python
sns.kdeplot(server.Time, cumulative = True)
```

```python
sns.kdeplot(mlb.Salary, cumulative = True)
```
